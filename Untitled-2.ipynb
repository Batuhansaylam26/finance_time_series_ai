```python
import yfinance as yf
import pandas as pd 
tickers = yf.download('msft aapl goog',period="5y")
ticker=['msft aapl goog']
```

```python
tickers["Close"]
```

```python
prices=pd.DataFrame(tickers["Close"])
```

```python
prices
```

```python
monthly_prices = prices.resample('M').last()
```

```python
"""monthly_prices"""
delta=prices.diff()
delta.iloc[0,:]=0
delta
```

```python
delta.columns.tolist()
```

```python


period = 14
for i in delta.columns.tolist():
    # Calculate the average gain and average loss for the specified period
    gain = pd.DataFrame(delta.loc[:,i]).where(delta > 0, 0)
    loss = -pd.DataFrame(delta.loc[:,i]).where(delta < 0, 0)
    avg_gain = gain.rolling(period).mean().dropna()
    avg_loss = loss.rolling(period).mean().dropna()

    # Calculate the Relative Strength (RS) by dividing the average gain by the average loss
    rs = avg_gain / avg_loss

    # Calculate the Relative Strength Index (RSI)
    rsi = 100 - (100 / (1 + rs))
    colname='RSİ_'+i
    # Add the RSI to the dataframe
    prices[colname] = rsi
# Print the dataframe
print(prices.dropna())
```

```python
outlier_cutoff = 0.01
data = pd.DataFrame()
lags = [1, 2, 3, 6, 9, 12]
for lag in lags:
    data[f'return_{lag}m'] = (monthly_prices
                            .pct_change(lag)
                            .stack()
                            .pipe(lambda x:
x.clip(lower=x.quantile(outlier_cutoff),
                        upper=x.quantile(1-outlier_cutoff)))
                            .add(1)
                            .pow(1/lag)
                            .sub(1)
                        )
data = data.swaplevel().dropna()
data
```

```python
for lag in [2,3,6,9,12]:
    data[f'momentum_{lag}'] = data[f'return_{lag}m'].sub(data.return_1m)
data[f'momentum_3_12'] = data[f'return_12m'].sub(data.return_3m)
```

```python
data.index
```

```python
for t in range(1, 7):
    data[f'return_1m_t-{t}'] =data.groupby(['tickers']).return_1m.shift(t)
```
